### Cleaning Data

Dataframe Day

```python
#Delete the holiday column because it is already represented in the weekday column.
day_df.drop(['holiday'], axis=1, inplace=True)
```

```python
#Change the data type of the date column from object to datetime
datetime_columns = ['dteday']

for column in datetime_columns:
  day_df[column] = pd.to_datetime(day_df[column])
day_df.info()
```

```python
#These seasonal changes represent: 1 is spring, 2 is summer, 3 is fall, 4 is winter
def change_season(x):
    if x == 1:
        return 'spring'
    elif x == 2:
        return 'summer'
    elif x == 3:
        return 'fall'
    else:
        return 'winter'

day_df['season'] = day_df['season'].apply(change_season)
day_df.head()
```

```python
#Change the year to represent: 0 is a 2011, 1 is a 2012
def change_year(x):
    if x == 0:
        return 2011
    else:
        return 2012

day_df['yr'] = day_df['yr'].apply(change_year)
day_df.head()
```

```python
#Change the month to represent: 1 is a January, 2 is a february, 3 is a march, 4 is a april, 5 is a may, 6 is a june, 7 is a july, 8 is a august, 9 is a september, 10 is a october, 11 is a november, 12 is a december
month_change = {1: 'january', 2: 'february', 3: 'march', 4:'april', 5:'may', 6:'june', 7:'july',
                8:'august', 9:'september', 10:'october', 11:'november', 12:'december'}

day_df['mnth'] = day_df['mnth'].map(month_change)
day_df.head()
```

```python
#Change the day to represent: 0 is a monday, 1 is a tuesday, 2 is a wednesday, 3 is a thursday, 4 is a friday, 5 is a saturday, 6 is a sunday
day_change = {0: 'monday', 1: 'tuesday', 2: 'wednesday', 3: 'thursday', 4:'friday', 5:'saturday', 6:'sunday'}

day_df['weekday'] = day_df['weekday'].map(day_change)
day_df.head(5)
```

```python
# Changing workingday data from numbers to weekend and workingday
def change_workingday(x):
    if x == 0:
        return 'weekend'
    else:
        return 'working day'

day_df['workingday'] = day_df['workingday'].apply(change_workingday)
day_df.head()
```

```python
#Change the weather to represent: 1 is a clear, 2 is a mist, 3 is a light snow, 4 is a heavy rain
def change_weather(x):
    if x == 1:
        return 'clear'
    elif x == 2:
        return 'mist'
    elif x == 3:
        return 'light snow'
    else:
        return 'heavy rain'

day_df['weathersit'] = day_df['weathersit'].apply(change_weather)
day_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: How has the company's sales performance been in recent years?

```python
monthly_counts = day_df.groupby(day_df['dteday'].dt.to_period('M'))['cnt'].max().reset_index()

fig, ax = plt.subplots(figsize=(24, 5))

sns.pointplot(data=monthly_counts, x='dteday', y='cnt', ax=ax, color="#90CAF9", marker='o', err_kws={'alpha': 0})  # Marker 'o' dan warna biru

ax.set_xlabel('Month')
ax.set_ylabel('Sum')
ax.set_title('Graph of Number of Customers per Month in 2012')

plt.show()
```

Based on the graph above, it can be seen that the highest number of orders occurred in September. Apart from that, there was also a quite drastic decline in the number of orders in November and December.

### Pertanyaan 2: How effective is bike sharing over a 24-hour period?

```python
fig, ax = plt.subplots(figsize=(25,10))
sns.pointplot(data=hour_df, x='hr', y='cnt', hue='workingday', errorbar=None, ax=ax, palette=['red', 'blue'])
ax.set(title='Bike Sharing over a 24-hour period')
ax.set_ylabel('cnt')
ax.set_xlabel('hr')
plt.show()
```

On weekdays, the number of users experiences a significant increase at the time of leaving and returning from work, namely 06.00 am to 08.00 am in the morning and 15.00 pm to 17.00 pm in the afternoon. A significant decrease in the number of users occurs at 09.00 am when people have left for work and at night starting at 19.00 pm. The number of users starts to decrease from 20.00 pm to 5 am.
On weekends, the number of users starts to increase starting at 06.00 am and continues to increase until it reaches its peak at 13.00 pm. After 16.00 pm, the number of users starts to decrease until 05.00.
